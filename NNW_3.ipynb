```python
# Zu Klassen siehe https://docs.python.org/3/tutorial/classes.html
# und insbesondere https://docs.python.org/3/tutorial/classes.html#tut-private
class myclass:
    def __init__(self,W): # Konstruktor, self entspricht this in Java/C++
        self._W = W # Konvention: private Variablen beginnen mit _
        # self muss beim Zugriff angegeben werden, sonst
        # ist es eine lokale Variable
    def f(self): # self muss immer explizit bergeben werden, sonst "static"
        return self._W*2
```

```python
m=myclass(np.array([-1,1]))
m._W # Zugriff ist auch auf private Variablen erlaubt
```

```python
m.f() # m wird automatisch als "self" bergeben
```

```python
import numpy as np
class SLN:
    def __init__(self, dIn, cOut):
        np.random.seed(42)
        self._W = np.random.randn(dIn, 1)
        self._W = self._W / np.sqrt(dIn+1)
        self._b = np.zeros((cOut,1))
        
    def neuron(self, X):
        net = np.zeros(X.shape[1])

        for n in range(X.shape[1]):
            net[n] = X[0,n] * self._W[0] + X[1,n] * self._W[1] - self._b
        return net>0
```

```python
sln = SLN(1, 1)
sln._W = np.array([-.3,1], ndmin=2).T
sln._b = np.array([2], ndmin=2)
nnwplot.plotTwoFeatures(X, T, sln.neuron)
```

```python
def ErrorRate(Y,T):
    if Y.ndim==1 or Y.shape[0]==1:
        errors=Y!=T
        return errors.sum()/Y.size
    else: # fr mehrere Ausgaben in one-hot Kodierung:
        # Dies brauchen Sie jetzt noch nicht nachzuvollziehen.
        errors=Y.argmax(0)!=T.argmax(0)
        return errors.sum()/Y.shape[1]
```
