```python
from sklearn.cluster import KMeans
clustering = KMeans(n_clusters=3)
X=iris[:,[0,3]]
clustering.fit(X)
nnwplot.plotTwoFeatures(X, T, clustering.predict)
```

```python
clustering.predict(X)
```

```python
clustering.fit_predict(X)
```

b) Bei mehrfachem Durchführen fällt auf, dass die Farben in der Grafik bzw die Klassennummern in den Arrays "durchtauschen". Das liegt daran, dass der Algorithmus mit k zufälligen DP anfängt, die Cluster aufzubauen.

c) die Farben bzw. die Label in T haben keine Bedeutung, da sie im Training gar nicht übergeben werden

```python
# d)
from sklearn.cluster import DBSCAN
from sklearn.neighbors import KNeighborsClassifier
clustering = DBSCAN(0.35)
y = clustering.fit_predict(X)
clf = KNeighborsClassifier(1)
clf.fit(X, y) # Kein Tippfehler, wirklich y - warum?
nnwplot.plotTwoFeatures(X, T, clf.predict)
```

DBSCAN kann nicht direkt zu plotTwoFeatures übergeben werden, da es keine predict() Methode gibt.
Deshalb wird ein Classifier auf die Cluster die bei DBSCAN entstehen trainiert. Der Classifier hat wieder eine predict Methode und der Plot kann gezeichnet werden.

## Aufgabe 3

```python
import matplotlib.pyplot as plt
#X = np.linspace(0, 2)
np.random.seed(42) # vereinfachte Variante, siehe np.random
X = np.sort(np.random.rand(50)*2)
T = np.sin(5*X) + X**2 + np.random.randn(len(X))/10
plt.plot(X, T, 'o')
plt.xlabel('x')
plt.ylabel('t')
plt.show()
```

```python
X.shape
```

a) 1 Merkmale &rarr; Merkmalraum ist 1dimensional

```python
plt.plot(X, T, 'o')
X = X.reshape(-1,1)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, T)

np.random.seed(91) # vereinfachte Variante, siehe np.random
y = reg.predict(X)
plt.plot(X, y, '-')
plt.show()
```

b) Sklearn erwartet als Merkmalseingabe eine Matrix mit der shape: (n_samples, n_features)

bei uns also: (50,1) &rarr; reshape macht das möglich

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X, T)
y = dtr.predict(X)
plt.plot(X, T, 'o')
plt.plot(X, y, '-')
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X, T)
y = svr.predict(X)
plt.plot(X, T, 'o')
plt.plot(X, y, '-')
```

```python
X = iris[:,:2]
T = iris[:,4]
svr.fit(X, T)
svr.predict(X)
nnwplot.plotTwoFeatures(X, T, svr.predict)
```

```python
decision = lambda x : np.round(svr.predict(x)) # eine kurze Funktionsdefinition
nnwplot.plotTwoFeatures(X, T, decision)
```
